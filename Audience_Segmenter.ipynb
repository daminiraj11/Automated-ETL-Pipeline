```python
import pandas as pd
from google.colab import files
import os

#step1:upload the file
uploaded=files.upload()
```

```python
#step2:get uploaded file dynamically
input_file=list(uploaded.keys())[0]

#step3:load excel file
df=pd.read_excel(input_file)

#step4:check columns
print(df.columns)
```

```python
#step5:check column names and update according to the target file column
segment_col='PRIORITY_SEGMENT_NEW'
npi_col='NPI_ID'
```

```python
#step6:filter required columns and drop missing NPI_ID
df=df[[segment_col,npi_col]].dropna(subset=[npi_col])

#step7:get unique segments
segments=df[segment_col].unique()

#step8:process each segment
for seg in segments:
  filtered=df[df[segment_col]==seg]

  output_df=filtered[[npi_col]].drop_duplicates()

  #Get unique NPI count
  npi_count=output_df.shape[0]

  safe_seg=str(seg).replace(" ","_")

  output_file=f"{safe_seg}_{npi_count}_{input_file}"
  output_df.to_excel(output_file,index=False)

  print(f"Saved:{output_file}")

#step9:Download generated files
for seg in segments:
  filtered=df[df[segment_col]==seg]

  output_df=filtered[[npi_col]].drop_duplicates()

  npi_count=output_df.shape[0]

  safe_seg=str(seg).replace(" ","_")
  output_file=f"{safe_seg}_{npi_count}_{input_file}"
  files.download(output_file)
```
